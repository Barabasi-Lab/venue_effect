```python
"""
Preprocess Biennale Art Data — unified pipeline for Venice + other biennials
════════════════════════════════════════════════════════════════════════════

INPUT FILES (place in ../../data/artist_info/):
  biennale_artist_match.csv    — Venice Biennale participants (artist_url, year)
  artfacts_artists_all.csv     — exploded exhibition records (all artists × exhibitions)
  bien_merge.csv               — other biennials participants

OUTPUT FILES (saved to ../../data/artist_info/ and ../../data/matching_needed/):
  treated_df_{venue_key}.parquet   — per-venue treated panels
  control_df_biennale.parquet      — shared control panel
"""

import warnings
from pathlib import Path

import numpy as np
import pandas as pd

warnings.filterwarnings('ignore')

# ══════════════════════════════════════════════════════════════════════
#  Configuration
# ══════════════════════════════════════════════════════════════════════
DATA_DIR   = Path('../../data')
INPUT_DIR  = DATA_DIR / 'artist_info'
OUTPUT_DIR = DATA_DIR / 'artist_info'
MATCH_DIR  = DATA_DIR / 'matching_needed'
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
MATCH_DIR.mkdir(parents=True, exist_ok=True)

PRE_WINDOW  = 5
POST_WINDOW = 10

VENUE_CONFIG = {
    'venice_biennale': {
        'institution_id': 2367, 'source': 'venice',
        'label': 'Venice Biennale',
    },
    'documenta': {
        'institution_id': 2538, 'source': 'other',
        'label': 'Documenta',
    },
    'bienal_sao_paulo': {
        'institution_id': 2549, 'source': 'other',
        'label': 'Bienal de São Paulo',
    },
    'whitney_biennial': {
        'institution_id': 5722, 'source': 'other',
        'label': 'The Whitney Biennial',
    },
    'biennale_of_sydney': {
        'institution_id': 4081, 'source': 'other',
        'label': 'Biennale of Sydney',
    },
    'istanbul_biennial': {
        'institution_id': 3933, 'source': 'other',
        'label': 'Istanbul Biennial',
    },
    'manifesta': {
        'institution_id': 6058, 'source': 'other',
        'label': 'Manifesta',
    },
    'gwangju_biennale': {
        'institution_id': 3458, 'source': 'other',
        'label': 'Gwangju Biennale',
    },
    'lyon_biennale': {
        'institution_id': 2493, 'source': 'other',
        'label': 'La Biennale de Lyon',
    },
    'berlin_biennale': {
        'institution_id': 2554, 'source': 'other',
        'label': 'Berlin Biennale',
    },
    'shanghai_biennale': {
        'institution_id': 3460, 'source': 'other',
        'label': 'Shanghai Biennale',
    },
    'sharjah_biennial': {
        'institution_id': 3462, 'source': 'other',
        'label': 'Sharjah Biennial',
    },
    'liverpool_biennial': {
        'institution_id': 4920, 'source': 'other',
        'label': 'Liverpool Biennial',
    },
    'taipei_biennial': {
        'institution_id': 6965, 'source': 'other',
        'label': 'Taipei Biennial',
    },
    'carnegie_international': {
        'institution_id': 41426, 'source': 'other',
        'label': 'Carnegie International',
    },
    'habana_bienal': {
        'institution_id': 4342, 'source': 'other',
        'label': 'Bienal de La Habana',
    },
    'yokohama_triennale': {
        'institution_id': 3457, 'source': 'other',
        'label': 'Yokohama Triennale',
    },
}

VENUES_TO_PROCESS = [
    'venice_biennale',
    'documenta',
    'bienal_sao_paulo',
    'whitney_biennial',
    'biennale_of_sydney',
    'istanbul_biennial',
    'manifesta',
    'gwangju_biennale',
]


# ══════════════════════════════════════════════════════════════════════
#  Country → Continent mapping
# ══════════════════════════════════════════════════════════════════════
COUNTRY_CONTINENT = {
    'Sierra Leone': 'Africa', 'Burkina Faso': 'Africa',
    'Saudi Arabia': 'Asia', 'Bolivia': 'South America',
    'Greenland': 'North America', 'Kiribati': 'Oceania',
    'Ukraine': 'Europe', 'Kenya': 'Africa', 'Thailand': 'Asia',
    'Uzbekistan': 'Asia', 'Bhutan': 'Asia', 'Kosovo': 'Europe',
    'Mongolia': 'Asia', 'Puerto Rico': 'North America',
    'Latvia': 'Europe', 'Togo': 'Africa', 'Jordan': 'Asia',
    'Germany': 'Europe', 'Barbados': 'North America',
    'Canada': 'North America', 'Colombia': 'South America',
    'Zambia': 'Africa', 'Belarus': 'Europe', 'Slovenia': 'Europe',
    'Italy': 'Europe', 'Venezuela': 'South America',
    'Brunei': 'Asia', 'Albania': 'Europe', 'Montenegro': 'Europe',
    'Botswana': 'Africa', 'Azerbaijan': 'Asia',
    'Ireland (Republic)': 'Europe', 'Pakistan': 'Asia',
    'Luxembourg': 'Europe', 'Liberia': 'Africa',
    'Cayman Islands': 'North America', 'China': 'Asia',
    'Denmark': 'Europe', 'Central African Republic': 'Africa',
    'Paraguay': 'South America', 'Namibia': 'Africa',
    'Liechtenstein': 'Europe', 'Portugal': 'Europe',
    'Dominica': 'North America', 'South Korea': 'Asia',
    'Malawi': 'Africa', 'Guinea-Bissau': 'Africa',
    'South Africa': 'Africa', 'Afghanistan': 'Asia',
    'Ethiopia': 'Africa', 'Algeria': 'Africa', 'Nigeria': 'Africa',
    'Myanmar': 'Asia', 'Equatorial Guinea': 'Africa',
    'Tajikistan': 'Asia', 'Kazakhstan': 'Asia',
    'Brazil': 'South America', 'Uruguay': 'South America',
    'Morocco': 'Africa', 'Swaziland': 'Africa', 'Eritrea': 'Africa',
    'Croatia': 'Europe', 'Romania': 'Europe', 'Tonga': 'Oceania',
    'Cameroon': 'Africa', 'North Macedonia': 'Europe',
    'Gambia': 'Africa', 'Iran': 'Asia', 'Mozambique': 'Africa',
    'Curaçao': 'North America', 'Bahamas': 'North America',
    'USA': 'North America', 'Bahrain': 'Asia',
    'Dominican Republic': 'North America',
    'Palestinian territories': 'Asia', 'São Tomé e Príncipe': 'Africa',
    'Kurdistan': 'Asia', 'United Arab Emirates': 'Asia',
    'Madagascar': 'Africa', 'Malaysia': 'Asia',
    'Czech Republic': 'Europe', 'Sudan': 'Africa',
    'Slovakia': 'Europe', 'Cyprus': 'Europe', 'Moldova': 'Europe',
    'Bulgaria': 'Europe', 'Lebanon': 'Asia', 'Oman': 'Asia',
    'Grenada': 'North America', 'Kuwait': 'Asia',
    'Mauritania': 'Africa', 'Lithuania': 'Europe',
    'San Marino': 'Europe', 'Micronesia': 'Oceania',
    'Rwanda': 'Africa', 'Spain': 'Europe',
    'Guatemala': 'North America', 'Taiwan': 'Asia',
    'Trinidad & Tobago': 'North America', 'Sri Lanka': 'Asia',
    'El Salvador': 'North America', 'Benin': 'Africa',
    'Hungary': 'Europe', 'Indonesia': 'Asia',
    'United States Virgin Islands': 'North America',
    'Malta': 'Europe', 'Bermuda': 'North America', 'India': 'Asia',
    'Australia': 'Oceania', 'Russia': 'Europe', 'Austria': 'Europe',
    'Libya': 'Africa', 'Seychelles': 'Africa', 'Vanuatu': 'Oceania',
    'Bosnia and Herzegovina': 'Europe', 'Angola': 'Africa',
    'Laos': 'Asia', 'Cape Verde': 'Africa',
    'Nicaragua': 'North America', 'Burundi': 'Africa',
    'New Zealand': 'Oceania', 'Finland': 'Europe',
    'Cuba': 'North America', 'Iraq': 'Asia',
    'Jamaica': 'North America', 'Zimbabwe': 'Africa',
    'Georgia': 'Asia', 'Faeroe Islands': 'Europe',
    'Egypt': 'Africa', 'Mali': 'Africa', 'Netherlands': 'Europe',
    'Vatican City State': 'Europe',
    'Netherlands Antilles': 'North America',
    'Philippines': 'Asia', 'Peru': 'South America',
    'Chile': 'South America', 'Nepal': 'Asia',
    'Ivory Coast': 'Africa', 'Chad': 'Africa', 'Belgium': 'Europe',
    'Vietnam': 'Asia', 'Greece': 'Europe', 'Monaco': 'Europe',
    'France': 'Europe', 'Haiti': 'North America',
    'Ecuador': 'South America', 'Senegal': 'Africa',
    'Bangladesh': 'Asia', 'Israel': 'Asia', 'Tanzania': 'Africa',
    'Papua New Guinea': 'Oceania', 'Mexico': 'North America',
    'Niger': 'Africa', 'Hong Kong': 'Asia', 'Uganda': 'Africa',
    'Iceland': 'Europe', 'Serbia': 'Europe',
    'Aruba': 'North America', 'Suriname': 'South America',
    'Estonia': 'Europe', 'United Kingdom': 'Europe',
    'Honduras': 'North America', 'Samoa': 'Oceania',
    'Qatar': 'Asia', 'Saint Barthélemy': 'North America',
    'Gabon': 'Africa', 'Argentina': 'South America',
    'Andorra': 'Europe', 'Singapore': 'Asia',
    'Reunion': 'Africa', 'East Timor': 'Asia',
    'Belize': 'North America', 'Costa Rica': 'North America',
    'Ghana': 'Africa', 'Tunisia': 'Africa',
    'Guyana': 'South America', 'Japan': 'Asia', 'Cambodia': 'Asia',
    'Saint Vincent and the Grenadines': 'North America',
    'Kyrgyzstan': 'Asia', 'Syria': 'Asia',
    'Turkey': 'Europe',   # Convention: map to Europe
    'Democratic Republic of the Congo': 'Africa',
    'Poland': 'Europe', 'Armenia': 'Europe',
    'Switzerland': 'Europe', 'North Korea': 'Asia',
    'Norway': 'Europe', 'Panama': 'North America',
    'Guadeloupe': 'North America', 'Sweden': 'Europe',
    'United States': 'North America', 'Mauritius': 'Africa',
    'Fiji Islands': 'Oceania',
    # Additional common entries
    'Ireland': 'Europe', 'Republic of Korea': 'Asia',
    'Martinique': 'North America', 'Guam': 'Oceania',
    'Yemen': 'Asia', 'Congo': 'Africa',
    'Republic of the Congo': 'Africa', 'Somalia': 'Africa',
    'Macao': 'Asia', 'Macau': 'Asia',
}

CONTINENT_COLS = ['Europe', 'North America', 'South America',
                  'Asia', 'Oceania', 'Africa']


def normalize_nationality(nat):
    """
    Ensure nationality is a flat list of strings.
    Handles:  list, "['A','B']", "A, B", "A", NaN, "None", "[]"
    """
    import ast
    if isinstance(nat, list):
        return [str(x).strip() for x in nat if x and str(x).strip()]
    if pd.isna(nat) or str(nat).strip() in ('None', '[]', 'nan', ''):
        return []
    s = str(nat).strip()
    # String repr of a list:  "['Germany', 'USA']"
    if s.startswith('['):
        try:
            parsed = ast.literal_eval(s)
            if isinstance(parsed, list):
                return [str(x).strip() for x in parsed if x]
        except Exception:
            pass
    # Comma-separated:  "Germany, USA"
    if ',' in s:
        return [x.strip() for x in s.split(',') if x.strip()]
    return [s]


def nationality_to_continent(nat_list):
    """Convert a list of nationalities → binary continent columns."""
    continents = {c: 0 for c in CONTINENT_COLS}
    for nat in nat_list:
        c = COUNTRY_CONTINENT.get(nat)
        if c and c in continents:
            continents[c] = 1
        elif c and '/' in c:
            # e.g. 'Europe/Asia' — set both
            for part in c.split('/'):
                part = part.strip()
                if part in continents:
                    continents[part] = 1
    return pd.Series(continents)


# ══════════════════════════════════════════════════════════════════════
#  Helper functions
# ══════════════════════════════════════════════════════════════════════

def categorize_career_stage(min_year, event_year):
    ca = event_year - min_year
    if ca < 10:   return 'early'
    elif ca < 30: return 'mid'
    else:         return 'late'

def categorize_age_stage(birth_year, event_year):
    if pd.isna(birth_year): return 'unknown'
    age = event_year - birth_year
    if age < 20:    return 'teen'
    elif age <= 39:  return 'adult'
    elif age <= 59:  return 'middle-age'
    elif age <= 100: return 'senior'
    else:            return 'posthumous'

def categorize_decade(year):
    return f"{(int(year) // 10) * 10}s"


def build_panel(type_counts_subset, label=''):
    """type_counts → full panel with cumulative S/G/F/B, min_year, career_age."""
    tc = type_counts_subset.copy()
    print(f"    Building panel for {label} "
          f"({tc['artist_id'].nunique():,} artists)...")

    pivot = (
        tc.pivot_table(
            index=['artist_id', 'end_year'], columns='type',
            values='count', aggfunc='sum', fill_value=0,
        ).reset_index()
    )
    for col in ['B', 'F', 'G', 'S']:
        if col not in pivot.columns:
            pivot[col] = 0

    artist_years = pivot.groupby('artist_id')['end_year'].agg(['min', 'max'])
    all_years = (
        artist_years
        .apply(lambda x: pd.Series(range(int(x['min']), int(x['max']) + 1)),
               axis=1)
        .stack().reset_index(level=1, drop=True)
        .reset_index(name='end_year')
    )
    all_years['end_year'] = all_years['end_year'].astype(int)

    panel = all_years.merge(
        pivot, on=['artist_id', 'end_year'], how='left'
    ).fillna(0)
    for col in ['B', 'F', 'G', 'S']:
        panel[col] = panel.groupby('artist_id')[col].cumsum()

    min_yr = panel.groupby('artist_id')['end_year'].min().rename('min_year')
    panel = panel.merge(min_yr, on='artist_id')
    panel['career_age'] = panel['end_year'] - panel['min_year']
    panel = panel.sort_values(['artist_id', 'end_year']).reset_index(drop=True)

    print(f"    → {panel['artist_id'].nunique():,} artists, "
          f"{len(panel):,} rows")
    return panel


# ══════════════════════════════════════════════════════════════════════
#  STEP 1: Load shared data
# ══════════════════════════════════════════════════════════════════════
print("=" * 70)
print("STEP 1: Loading shared data")
print("=" * 70)

# ── All exhibitions ──
exh_path = INPUT_DIR / 'artfacts_artists_all.csv'
if not exh_path.exists():
    exh_path = INPUT_DIR / 'artfacts_artists_all_good.csv'
print(f"\n  Exhibitions: {exh_path}")
all_exh = pd.read_csv(exh_path, low_memory=False)

if 'end_year' not in all_exh.columns:
    all_exh['end_year'] = all_exh['end_date'].apply(
        lambda x: int(str(x).split('-')[0]) if pd.notna(x) else np.nan
    )
all_exh = all_exh.dropna(subset=['end_year'])
all_exh['end_year'] = all_exh['end_year'].astype(int)
all_exh['artist_id'] = all_exh['artist_id'].astype(int)
print(f"  → {len(all_exh):,} rows, {all_exh['artist_id'].nunique():,} artists")

# ── Type counts ──
print("\n  Building type counts...")
type_counts = (
    all_exh
    .groupby(['artist_name', 'artist_id', 'end_year', 'type'])
    .size().reset_index(name='count')
)
print(f"  → {len(type_counts):,} rows")


# ══════════════════════════════════════════════════════════════════════
#  STEP 2: Build artist demographics with continent columns
# ══════════════════════════════════════════════════════════════════════
print("\n" + "=" * 70)
print("STEP 2: Building artist demographics (name, birth year, continents)")
print("=" * 70)

# Check if continent columns already exist
has_continents = all(c in all_exh.columns for c in CONTINENT_COLS)

if has_continents:
    print("  Continent columns already present in exhibition data.")
    # Even if present, an artist may have different flags across rows
    # (e.g., exhibited in different countries over time).
    # Take the MAX per artist so that any 1 survives.
    demo_cols = (['artist_id', 'artist_name', 'artist_birth_year']
                 + CONTINENT_COLS)
    available = [c for c in demo_cols if c in all_exh.columns]
    agg_dict = {}
    for c in available:
        if c == 'artist_id':
            continue
        elif c in CONTINENT_COLS:
            agg_dict[c] = 'max'       # 1 if ANY row has 1
        elif c == 'artist_name':
            agg_dict[c] = 'first'
        elif c == 'artist_birth_year':
            agg_dict[c] = 'first'
    artists_demo = (
        all_exh[available]
        .groupby('artist_id')
        .agg(agg_dict)
        .reset_index()
    )
else:
    print("  Continent columns NOT found — deriving from artist_nationality...")
    if 'artist_nationality' not in all_exh.columns:
        raise ValueError(
            "Neither continent columns nor artist_nationality found in data. "
            "Cannot derive regional information."
        )

    # Collect ALL nationality values per artist across all rows,
    # then union them so multi-nationality artists get all continents.
    print("  Aggregating nationalities per artist (union across all rows)...")
    nat_series = all_exh[['artist_id', 'artist_nationality']].copy()
    nat_series['artist_nationality'] = (
        nat_series['artist_nationality'].apply(normalize_nationality)
    )
    # Group by artist → flatten and deduplicate nationality lists
    artist_nat_agg = (
        nat_series
        .groupby('artist_id')['artist_nationality']
        .apply(lambda lists: list(set(
            n for sublist in lists for n in sublist if n
        )))
        .reset_index()
    )

    # Get name + birth year (first non-null)
    name_birth = (
        all_exh[['artist_id', 'artist_name', 'artist_birth_year']]
        .drop_duplicates(subset='artist_id')
    )
    artist_nat_agg = artist_nat_agg.merge(name_birth, on='artist_id', how='left')
    print(f"  Unique artists with nationality data: {len(artist_nat_agg):,}")

    # Map to continent binary columns
    print("  Mapping nationalities → continents...")
    continent_df = artist_nat_agg['artist_nationality'].apply(
        nationality_to_continent
    )
    artists_demo = pd.concat(
        [artist_nat_agg[['artist_id', 'artist_name', 'artist_birth_year']]
         .reset_index(drop=True),
         continent_df.reset_index(drop=True)],
        axis=1,
    )

    # Check coverage
    has_any = artists_demo[CONTINENT_COLS].sum(axis=1) > 0
    print(f"  Artists with ≥1 continent assigned: "
          f"{has_any.sum():,} / {len(artists_demo):,} "
          f"({has_any.mean():.1%})")

    # Show unmapped nationalities for debugging
    if has_any.mean() < 0.9:
        unmapped_mask = ~has_any
        unmapped_nats = (
            artist_nat.loc[unmapped_mask.values, 'artist_nationality']
            .explode().dropna()
        )
        unmapped_unique = unmapped_nats[
            ~unmapped_nats.isin(COUNTRY_CONTINENT.keys())
        ].value_counts().head(20)
        if len(unmapped_unique) > 0:
            print(f"  Top unmapped nationalities:")
            for nat, cnt in unmapped_unique.items():
                print(f"    {nat}: {cnt}")

print(f"  Demographics table: {len(artists_demo):,} artists, "
      f"cols: {list(artists_demo.columns)}")

# Save demographics for reuse
artists_demo.to_csv(
    OUTPUT_DIR / 'artists_continents.csv', index=False
)
print(f"  ✓ Saved artists_continents.csv")


# ══════════════════════════════════════════════════════════════════════
#  STEP 3: Gender estimation
# ══════════════════════════════════════════════════════════════════════
print("\n" + "=" * 70)
print("STEP 3: Gender estimation")
print("=" * 70)

try:
    import gender_guesser.detector as gender
    detector = gender.Detector(case_sensitive=False)

    def _get_gender(name):
        if pd.isna(name): return 'unknown'
        return detector.get_gender(str(name).split()[0])

    artist_gender = artists_demo[['artist_id', 'artist_name']].copy()
    artist_gender['gender_est'] = (
        artist_gender['artist_name'].apply(_get_gender)
    )
    artist_gender = artist_gender[['artist_id', 'gender_est']]

    g_counts = artist_gender['gender_est'].value_counts()
    print(f"  Gender distribution: {g_counts.to_dict()}")
except ImportError:
    print("  WARNING: gender_guesser not installed — using 'unknown'")
    artist_gender = artists_demo[['artist_id']].copy()
    artist_gender['gender_est'] = 'unknown'


# ══════════════════════════════════════════════════════════════════════
#  STEP 4: Load venue participant lists
# ══════════════════════════════════════════════════════════════════════
print("\n" + "=" * 70)
print("STEP 4: Loading venue participant lists")
print("=" * 70)

# Venice
venice_path = INPUT_DIR / 'biennale_artist_match.csv'
print(f"\n  Venice: {venice_path}")
venice_artists = pd.read_csv(venice_path)
venice_artists['artist_id'] = venice_artists['artist_url'].apply(
    lambda x: int(str(x).split('/')[-1])
)
venice_artists['institution_id'] = 2367
print(f"  → {venice_artists['artist_id'].nunique():,} artists")

# Other biennials
bien_path = INPUT_DIR / 'bien_merge.csv'
other_artists = pd.DataFrame()
if bien_path.exists():
    print(f"\n  Other biennials: {bien_path}")
    bien_merge = pd.read_csv(bien_path, low_memory=False)
    bien_merge['artist_id'] = bien_merge['artist_url'].apply(
        lambda x: int(str(x).split('/')[-1])
    )
    if 'institution_id' not in bien_merge.columns:
        bien_merge['institution_id'] = bien_merge['venue_links_card'].apply(
            lambda x: int(str(x).split('/')[-1])
        )
    if 'year' not in bien_merge.columns:
        bien_merge['year'] = bien_merge['end_date'].apply(
            lambda x: int(str(x).split('-')[0]) if pd.notna(x) else np.nan
        )
    other_artists = bien_merge[['artist_id', 'institution_id', 'year']].dropna()
    other_artists['year'] = other_artists['year'].astype(int)
    print(f"  → {other_artists['artist_id'].nunique():,} artists, "
          f"{other_artists['institution_id'].nunique()} venues")
else:
    print(f"  WARNING: {bien_path} not found — only Venice")


# ══════════════════════════════════════════════════════════════════════
#  STEP 5: Build shared CONTROL panel
#
#  Control excludes ONLY top-tier venues (Venice + Documenta).
#  Lower-tier venue artists remain in the control pool here;
#  at matching time, each lower-tier venue will additionally
#  exclude its own artists using the full participant ID lists
#  saved below.
# ══════════════════════════════════════════════════════════════════════
print("\n" + "=" * 70)
print("STEP 5: Building shared CONTROL panel")
print("=" * 70)

# Top-tier venue IDs to exclude from control
TOP_TIER_IDS = {2367, 2538}  # Venice Biennale, Documenta
top_tier_artist_ids = set(venice_artists['artist_id'].unique())
if not other_artists.empty:
    documenta_ids = set(
        other_artists.loc[
            other_artists['institution_id'] == 2538, 'artist_id'
        ].unique()
    )
    top_tier_artist_ids |= documenta_ids
print(f"  Top-tier (Venice + Documenta) artists excluded: "
      f"{len(top_tier_artist_ids):,}")

# Save FULL participant ID lists per venue for later matching exclusion.
# These include ALL artists who ever appeared at each venue,
# not just the windowed/filtered treated panels.
venue_id_lists = {}
venue_id_lists['venice_biennale'] = set(venice_artists['artist_id'].unique())
if not other_artists.empty:
    for vk, vcfg in VENUE_CONFIG.items():
        if vcfg['source'] == 'venice':
            continue
        iid = vcfg['institution_id']
        ids = set(
            other_artists.loc[
                other_artists['institution_id'] == iid, 'artist_id'
            ].unique()
        )
        if ids:
            venue_id_lists[vk] = ids

# Save ID lists for matching step
import pickle
for dest in [OUTPUT_DIR, MATCH_DIR]:
    id_list_path = dest / 'venue_artist_id_lists.pkl'
    with open(id_list_path, 'wb') as f:
        pickle.dump(venue_id_lists, f)
    print(f"  ✓ Saved venue artist ID lists → {id_list_path}")
for vk, ids in venue_id_lists.items():
    print(f"    {vk}: {len(ids):,} artists")

# Build control: exclude only top-tier
control_tc = type_counts[~type_counts['artist_id'].isin(top_tier_artist_ids)]
control_panel = build_panel(control_tc, 'control')

# Attach demographics + gender
control_panel = control_panel.merge(artists_demo, on='artist_id', how='left')
if 'artist_birth_year' in control_panel.columns:
    control_panel['age'] = (
        control_panel['end_year'] - control_panel['artist_birth_year']
    )
control_panel = control_panel.merge(artist_gender, on='artist_id', how='left')

# Placeholder columns
control_panel['biennale_year']    = np.nan
control_panel['biennale_treated'] = 0
control_panel['year_diff']        = np.nan
control_panel['career_stage']     = np.nan
control_panel['age_stage']        = np.nan
control_panel['biennale_decade']  = np.nan
control_panel['institution_id']   = np.nan

print(f"  Control: {control_panel['artist_id'].nunique():,} artists, "
      f"{len(control_panel):,} rows")

# Save
for dest in [OUTPUT_DIR, MATCH_DIR]:
    cp = dest / 'control_df_biennale.parquet'
    control_panel.to_parquet(cp, index=False)
    print(f"  ✓ {cp}")
control_panel.to_csv(OUTPUT_DIR / 'control_df_biennale.csv', index=False)


# ══════════════════════════════════════════════════════════════════════
#  STEP 6: Process each venue
# ══════════════════════════════════════════════════════════════════════
for venue_key in VENUES_TO_PROCESS:
    cfg = VENUE_CONFIG[venue_key]
    inst_id = cfg['institution_id']
    label   = cfg['label']

    print("\n" + "=" * 70)
    print(f"VENUE: {label}  (id={inst_id}, key={venue_key})")
    print("=" * 70)

    # ── Get participants ──
    if cfg['source'] == 'venice':
        vp = venice_artists[['artist_id', 'year']].copy()
        vp['institution_id'] = inst_id
    else:
        if other_artists.empty:
            print(f"  SKIP: no other-biennials data")
            continue
        vp = other_artists[
            other_artists['institution_id'] == inst_id
        ][['artist_id', 'year', 'institution_id']].copy()

    if vp.empty:
        print(f"  SKIP: no participants for id={inst_id}")
        continue

    print(f"  Participants: {vp['artist_id'].nunique():,} artists")

    # ── First appearance per artist ──
    first_appear = (
        vp.groupby('artist_id')['year'].min()
        .reset_index().rename(columns={'year': 'biennale_year'})
    )
    first_appear['institution_id'] = inst_id

    # ── Build treated panel ──
    venue_ids = set(vp['artist_id'].unique())
    treated_tc = type_counts[type_counts['artist_id'].isin(venue_ids)]
    treated_panel = build_panel(treated_tc, f'treated ({venue_key})')

    # Demographics + gender
    treated_panel = treated_panel.merge(
        artists_demo, on='artist_id', how='left'
    )
    if 'artist_birth_year' in treated_panel.columns:
        treated_panel['age'] = (
            treated_panel['end_year'] - treated_panel['artist_birth_year']
        )
    treated_panel = treated_panel.merge(
        artist_gender, on='artist_id', how='left'
    )

    # Merge biennale_year
    treated_panel = treated_panel.merge(
        first_appear[['artist_id', 'biennale_year', 'institution_id']],
        on='artist_id', how='inner',
    )

    # Treatment indicator + year_diff
    treated_panel['biennale_treated'] = (
        treated_panel['end_year'] >= treated_panel['biennale_year']
    ).astype(int)
    treated_panel['year_diff'] = (
        treated_panel['end_year'] - treated_panel['biennale_year']
    )

    # Career stage, age stage, decade
    treated_panel['career_stage'] = treated_panel.apply(
        lambda r: categorize_career_stage(r['min_year'], r['biennale_year']),
        axis=1,
    )
    if 'artist_birth_year' in treated_panel.columns:
        treated_panel['age_stage'] = treated_panel.apply(
            lambda r: categorize_age_stage(
                r['artist_birth_year'], r['biennale_year']
            ), axis=1,
        )
    else:
        treated_panel['age_stage'] = 'unknown'
    treated_panel['biennale_decade'] = (
        treated_panel['biennale_year'].apply(categorize_decade)
    )

    # ── Filter to [-5, +10] window ──
    n_before = len(treated_panel)
    treated_panel = treated_panel[
        (treated_panel['year_diff'] >= -PRE_WINDOW) &
        (treated_panel['year_diff'] <= POST_WINDOW)
    ].copy()

    # Keep artists with sufficient coverage on both sides
    valid_ids = treated_panel.groupby('artist_id').filter(
        lambda x: (x['year_diff'].min() <= -PRE_WINDOW + 1) and
                  (x['year_diff'].max() >= POST_WINDOW - 1)
    )['artist_id'].unique()
    treated_panel = treated_panel[
        treated_panel['artist_id'].isin(valid_ids)
    ].copy()

    print(f"  Window [-{PRE_WINDOW}, +{POST_WINDOW}]: "
          f"{treated_panel['artist_id'].nunique():,} artists, "
          f"{len(treated_panel):,} rows (from {n_before:,})")

    # ── Sanity checks ──
    overlap = set(treated_panel['artist_id']) & set(control_panel['artist_id'])
    dupes = treated_panel.groupby(['artist_id', 'end_year']).size()
    print(f"  Overlap w/ control: {len(overlap)} (expect 0)")
    if len(dupes) > 0:
        print(f"  Max rows/(artist,year): {dupes.max()}")

    # ── Save ──
    for dest in [OUTPUT_DIR, MATCH_DIR]:
        tp = dest / f'treated_df_{venue_key}.parquet'
        treated_panel.to_parquet(tp, index=False)
        print(f"  ✓ {tp}")
    treated_panel.to_csv(
        OUTPUT_DIR / f'treated_df_{venue_key}.csv', index=False
    )

    # Summary
    print(f"\n  {label} summary:")
    print(f"    Artists: {treated_panel['artist_id'].nunique():,}")
    print(f"    Columns: {list(treated_panel.columns)}")
    if 'gender_est' in treated_panel.columns:
        g = treated_panel.drop_duplicates('artist_id')['gender_est'].value_counts()
        print(f"    Gender: {g.head(5).to_dict()}")
    cs = treated_panel.drop_duplicates('artist_id')['career_stage'].value_counts()
    print(f"    Career stage: {cs.to_dict()}")
    for c in CONTINENT_COLS:
        if c in treated_panel.columns:
            n = treated_panel.drop_duplicates('artist_id')[c].sum()
            if n > 0:
                print(f"    {c}: {int(n)}")


# ══════════════════════════════════════════════════════════════════════
print("\n" + "=" * 70)
print("ALL DONE")
print("=" * 70)
print(f"\nOutputs in:\n  {OUTPUT_DIR}\n  {MATCH_DIR}")
print(f"\nFiles:")
for p in sorted(OUTPUT_DIR.glob('*_df_*.parquet')) + \
         sorted(OUTPUT_DIR.glob('*_df_*.csv')):
    print(f"  {p.name}  ({p.stat().st_size / 1e6:.1f} MB)")
```
