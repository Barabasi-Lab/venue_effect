# Art Biennale — Venue Effect Plots

1. General curves (year_diff = -5 to 10): treated vs control for S, G, F, B
2. Decade cohort plots — single venue
3. Summary table
4. **Decade cohort plots — Venice + Documenta (shared axes per metric)**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# ── Plotting defaults ──────────────────────────────────────────
plt.rcParams.update({
    'axes.spines.top':    False,
    'axes.spines.right':  False,
    'axes.spines.left':   False,
    'ytick.left':         False,
    'axes.grid':          True,
    'axes.grid.which':    'major',
    'axes.grid.axis':     'y',
    'grid.linestyle':     '--',
    'grid.alpha':         0.3,
})
```

## Configuration

```python
# ══════════════════════════════════════════════════════════════
#  EDIT THESE
# ══════════════════════════════════════════════════════════════
VENUE_KEY  = 'venice_biennale'
DATA_PATH  = Path('../../data/matches/matched_venice_biennale.parquet')
# DATA_PATH  = Path('../../data/matches//enriched_titles/matched_documenta_enriched_titles.csv')

SAVE_FIGS  = False       # ← toggle True to write files
OUTPUT_DIR = Path('../../figure/art')

MIN_YEAR   = 1895           # earliest biennale_year to include in decade plots
POST_YEAR  = 5              # which year_diff to use for decade cohort outcome

# Metrics to plot
METRICS = [
    ('S',     '# Solo exhibitions'),
    ('G',     '# Group exhibitions'),
    ('F',     '# Art fairs'),
    ('B_adj', '# Biennales (adj)'),
    ('S_titles', '# Pre-venue titles'),
    # ('S_titles_llm', '{re-venue titles LLMs')
]

# Colors
COLOR_TREATED = 'wheat'         # general curves: treated
COLOR_CONTROL = 'darkgray'      # general curves: control

# Decade palette
DECADE_PALETTE = plt.cm.Set3_r

print(f'Venue: {VENUE_KEY}')
print(f'Data:  {DATA_PATH}')
print(f'Save:  {SAVE_FIGS}')
```

## Load data

```python
if str(DATA_PATH).endswith('.parquet'):
    df = pd.read_parquet(DATA_PATH)
elif str(DATA_PATH).endswith('.csv'):
    # Try semicolon first (science format), then comma
    try:
        df = pd.read_csv(DATA_PATH, sep=';')
        if len(df.columns) < 3:
            df = pd.read_csv(DATA_PATH, sep=',')
    except:
        df = pd.read_csv(DATA_PATH, sep=',')
else:
    df = pd.read_csv(DATA_PATH)

print(f'Loaded: {len(df):,} rows, {df["artist_id"].nunique():,} artists')
print(f'Columns: {list(df.columns)}')
df.head(3)
```

```python
# ── Split treated / control ──
if 'matched_to' in df.columns:
    # treated: matched_to == self (artist_id)
    treated_ids = set(
        df.loc[df['matched_to'] == df['artist_id'], 'artist_id'].unique()
    )
    control_ids = set(df['artist_id'].unique()) - treated_ids
elif 'is_venue' in df.columns:
    treated_ids = set(df.loc[df['is_venue'] == 1, 'artist_id'].unique())
    control_ids = set(df['artist_id'].unique()) - treated_ids
else:
    # Fallback: biennale_treated
    treated_ids = set(
        df.loc[df['biennale_treated'] == 1, 'artist_id'].unique()
    )
    control_ids = set(df['artist_id'].unique()) - treated_ids

print(f'Treated: {len(treated_ids):,}  Control: {len(control_ids):,}')
```

## Helper functions

```python
def compute_stats(df, col, artist_ids, time_col='year_diff'):
    """Per time-point: mean, SE, 95% CI, n."""
    sub = df[df['artist_id'].isin(artist_ids)].copy()
    sub[col] = pd.to_numeric(sub[col], errors='coerce')
    rows = []
    for ty in sorted(sub[time_col].dropna().unique()):
        vals = sub.loc[sub[time_col] == ty, col].dropna()
        n = len(vals)
        if n == 0:
            continue
        m = vals.mean()
        se = vals.std(ddof=1) / np.sqrt(n) if n > 1 else 0
        rows.append({'time': ty, 'mean': m, 'se': se, 'n': n,
                     'ci_lo': m - 1.96 * se, 'ci_hi': m + 1.96 * se})
    return pd.DataFrame(rows)


def get_post_stats_by_decade(df, col, artist_ids, post_year=5, min_year=1895):
    """For each decade, mean +/- 95% CI of `col` at year_diff == post_year."""
    sub = df[
        (df['artist_id'].isin(artist_ids)) &
        (df['year_diff'] == post_year)
    ].copy()
    sub = sub[sub['biennale_year'] >= min_year]
    sub[col] = pd.to_numeric(sub[col], errors='coerce')
    sub = sub.dropna(subset=[col, 'biennale_year'])
    sub['decade'] = (sub['biennale_year'] // 10 * 10).astype(int)

    rows = []
    for dec, grp in sub.groupby('decade'):
        vals = grp[col].dropna()
        n = len(vals)
        if n == 0:
            continue
        m = vals.mean()
        se = vals.std(ddof=1) / np.sqrt(n) if n > 1 else 0
        rows.append({'decade': dec, 'mean': m, 'se': se, 'n': n,
                     'ci_lo': m - 1.96 * se, 'ci_hi': m + 1.96 * se})
    return pd.DataFrame(rows).sort_values('decade').reset_index(drop=True)
```

## 1 · General curves (year_diff = -5 to 10)

```python
def plot_general(t_stats, c_stats, ylabel, title=None, out_path=None):
    """Treated vs control trajectory with 95% CI."""
    ax_w, ax_h = 2, 2
    lm, bm, rm, tm = 0.75, 0.55, 0.15, 0.25
    fw, fh = lm + ax_w + rm, bm + ax_h + tm
    fig = plt.figure(figsize=(fw, fh), dpi=300)
    fig.patch.set_alpha(0)
    ax = fig.add_axes([lm/fw, bm/fh, ax_w/fw, ax_h/fh])
    ax.patch.set_alpha(0)

    # Control
    ax.errorbar(c_stats['time'], c_stats['mean'],
                yerr=1.96 * c_stats['se'],
                color=COLOR_CONTROL, fmt='none',
                capsize=3, capthick=0.96, elinewidth=0.96)
    ax.plot(c_stats['time'], c_stats['mean'],
            '-o', color=COLOR_CONTROL, markersize=4, linewidth=2,
            alpha=0.5, label='Control')

    # Treated
    ax.errorbar(t_stats['time'], t_stats['mean'],
                yerr=1.96 * t_stats['se'],
                color=COLOR_TREATED, fmt='none',
                capsize=3, capthick=0.96, elinewidth=0.96)
    ax.plot(t_stats['time'], t_stats['mean'],
            '-o', color=COLOR_TREATED, markersize=4, linewidth=2,
            alpha=0.8, label='Venue access')

    ax.axvline(x=0, color='grey', linestyle='--', linewidth=1.5)
    ax.set_xlabel('Time to participate')
    ax.set_ylabel(ylabel)
    ax.set_xlim(-5.5, 10.5)
    if title:
        ax.set_title(title, fontsize=9, pad=4)

    if out_path and SAVE_FIGS:
        out_path = Path(out_path)
        out_path.parent.mkdir(parents=True, exist_ok=True)
        fig.savefig(out_path, dpi=600, transparent=True, bbox_inches='tight')
        fig.savefig(out_path.with_suffix('.pdf'), bbox_inches='tight', transparent=True)
        print(f'  Saved → {out_path}')
        plt.close(fig)
    else:
        plt.show()
```

```python
if SAVE_FIGS:
    OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

for col, ylabel in METRICS:
    if col not in df.columns:
        print(f'  ⚠ {col} not found — skipping')
        continue
    print(ylabel)
    # print(c_stats)
    t_stats = compute_stats(df, col, treated_ids)
    c_stats = compute_stats(df, col, control_ids)

    out_path = OUTPUT_DIR / f'{VENUE_KEY}_{col}_general.png'
    plot_general(t_stats, c_stats, ylabel,
                 out_path=out_path)
```

```python
if SAVE_FIGS:
    OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

for col, ylabel in METRICS:
    if col not in df.columns:
        print(f'  ⚠ {col} not found — skipping')
        continue
    print(ylabel)
    # print(c_stats)
    t_stats = compute_stats(df, col, treated_ids)
    c_stats = compute_stats(df, col, control_ids)

    out_path = OUTPUT_DIR / f'{VENUE_KEY}_{col}_general.png'
    plot_general(t_stats, c_stats, ylabel,
                 out_path=out_path)
```

## 2 · Decade cohort plots — single venue

```python
def plot_decade(t_dec, c_dec, ylabel, color, title=None, out_path=None,
                y_lim=None, x_ticks=None, x_lim=None):
    """Decade cohort: treated vs control."""
    ax_w, ax_h = 3, 1.2
    lm, bm, rm, tm = 0.85, 0.55, 0.15, 0.30
    fw, fh = lm + ax_w + rm, bm + ax_h + tm
    fig = plt.figure(figsize=(fw, fh), dpi=300)
    ax = fig.add_axes([lm/fw, bm/fh, ax_w/fw, ax_h/fh])

    # Control
    if not c_dec.empty:
        ax.fill_between(c_dec['decade'], c_dec['ci_lo'], c_dec['ci_hi'],
                        color='darkgray', alpha=0.2)
        ax.plot(c_dec['decade'], c_dec['mean'],
                '-o', color='darkgray', linewidth=1.8, alpha=0.7,
                markersize=4, label='Control')

    # Treated
    if not t_dec.empty:
        ax.fill_between(t_dec['decade'], t_dec['ci_lo'], t_dec['ci_hi'],
                        color=color, alpha=0.2)
        ax.plot(t_dec['decade'], t_dec['mean'],
                '-o', color=color, markersize=4, linewidth=1.8, alpha=0.8,
                label='Treated')

    ax.set_xlabel('Decade', fontsize=8)
    ax.set_ylabel(ylabel, fontsize=8)
    ax.tick_params(labelsize=6)

    if x_ticks is not None:
        ax.set_xticks(x_ticks)
    if x_lim is not None:
        ax.set_xlim(x_lim)
    if y_lim is not None:
        ax.set_ylim(y_lim)
    if title:
        ax.set_title(title, fontsize=9, pad=4)

    if out_path and SAVE_FIGS:
        out_path = Path(out_path)
        out_path.parent.mkdir(parents=True, exist_ok=True)
        fig.savefig(out_path, dpi=600, bbox_inches='tight', transparent=True)
        fig.savefig(out_path.with_suffix('.pdf'), bbox_inches='tight', transparent=True)
        print(f'  Saved → {out_path}')
        plt.close(fig)
    else:
        plt.show()
```

## 3 · Summary table

```python
# ── Log all stats ──
log_rows = []

for col, ylabel in METRICS:
    if col not in df.columns:
        continue
    t_stats = compute_stats(df, col, treated_ids)
    c_stats = compute_stats(df, col, control_ids)

    for _, r in t_stats.iterrows():
        log_rows.append({
            'venue': VENUE_KEY, 'metric': col, 'group': 'treated',
            'year_diff': int(r['time']), 'mean': r['mean'],
            'ci_lo': r['ci_lo'], 'ci_hi': r['ci_hi'], 'n': int(r['n']),
        })
    for _, r in c_stats.iterrows():
        log_rows.append({
            'venue': VENUE_KEY, 'metric': col, 'group': 'control',
            'year_diff': int(r['time']), 'mean': r['mean'],
            'ci_lo': r['ci_lo'], 'ci_hi': r['ci_hi'], 'n': int(r['n']),
        })

log_df = pd.DataFrame(log_rows)
print(f'Log table: {len(log_df)} rows')
log_df.head(10)
```

## 4 · Multi-venue decade: Venice + Documenta (shared axes per metric)

For each metric (S, G, F, B_adj), plot Venice and Documenta side-by-side.
Same x-axis (decade range) and same y-axis (value range) so they are directly comparable.

```python
# ══════════════════════════════════════════════════════════════
#  Multi-venue config — EDIT THESE
# ══════════════════════════════════════════════════════════════
MULTI_VENUES = [
    ('venice_biennale', '../../data/matches/matched_venice_biennale.parquet', 'Venice Biennale'),
    ('documenta',       '../../data/matches/matched_documenta.parquet',       'Documenta'),
]

MULTI_MIN_YEAR = 1900
MULTI_POST_YEAR = POST_YEAR   # reuse from above
```

```python
# ── Pass 1: Load all venues, compute decade stats ──
multi_data = {}

for vkey, vpath, vlabel in MULTI_VENUES:
    vpath = Path(vpath)
    if not vpath.exists():
        print(f'  WARNING: {vpath} not found — skipping {vlabel}')
        continue

    if str(vpath).endswith('.parquet'):
        vdf = pd.read_parquet(vpath)
    else:
        vdf = pd.read_csv(vpath)
    print(f'{vlabel}: {len(vdf):,} rows, {vdf["artist_id"].nunique():,} artists')

    # Split treated / control
    if 'matched_to' in vdf.columns:
        t_ids = set(vdf.loc[vdf['matched_to'] == vdf['artist_id'], 'artist_id'].unique())
    elif 'is_venue' in vdf.columns:
        t_ids = set(vdf.loc[vdf['is_venue'] == 1, 'artist_id'].unique())
    else:
        t_ids = set(vdf.loc[vdf['biennale_treated'] == 1, 'artist_id'].unique())
    c_ids = set(vdf['artist_id'].unique()) - t_ids
    print(f'  Treated: {len(t_ids):,}  Control: {len(c_ids):,}')

    venue_metrics = {}
    for col, ylabel in METRICS:
        if col not in vdf.columns:
            continue
        t_dec = get_post_stats_by_decade(vdf, col, t_ids,
                                         post_year=MULTI_POST_YEAR,
                                         min_year=MULTI_MIN_YEAR)
        c_dec = get_post_stats_by_decade(vdf, col, c_ids,
                                         post_year=MULTI_POST_YEAR,
                                         min_year=MULTI_MIN_YEAR)
        venue_metrics[col] = {'ylabel': ylabel, 't_dec': t_dec, 'c_dec': c_dec}

    multi_data[vkey] = {'label': vlabel, 'metrics': venue_metrics}

print(f'\nLoaded {len(multi_data)} venues')
```

```python
# ── Pass 2: Compute shared y-limits and x-ticks per metric (across all venues) ──
shared_ylims = {}   # col -> (ymin, ymax)
shared_xticks = {}  # col -> list of decade ints
shared_xlim = {}    # col -> (xmin, xmax)

for col, ylabel in METRICS:
    ymins, ymaxs = [], []
    all_decs = []

    for vkey in multi_data:
        m = multi_data[vkey]['metrics'].get(col)
        if m is None:
            continue
        for key in ['t_dec', 'c_dec']:
            d = m[key]
            if not d.empty:
                ymins.append(d['ci_lo'].min())
                ymaxs.append(d['ci_hi'].max())
                all_decs.extend(d['decade'].tolist())

    if ymins and ymaxs:
        ylo, yhi = min(ymins), max(ymaxs)
        pad = 0.05 * (yhi - ylo) if yhi > ylo else 1
        shared_ylims[col] = (ylo - pad, yhi + pad)

    if all_decs:
        dmin = (min(all_decs) // 10) * 10
        dmax = max(all_decs)
        shared_xticks[col] = list(range(dmin, dmax + 1, 10))
        shared_xlim[col] = (dmin - 5, dmax + 5)

    print(f'{col}: y_lim={shared_ylims.get(col)}, '
          f'x_ticks={shared_xticks.get(col)}')
```

```python
# ── Pass 3: Separate plots per venue per metric (shared axes, color by venue) ──

venue_keys_ordered = [vkey for vkey, _, _ in MULTI_VENUES if vkey in multi_data]
n_venues = len(venue_keys_ordered)

# Color by VENUE (not by metric)
venue_colors = {}
for i, vkey in enumerate(venue_keys_ordered):
    venue_colors[vkey] = DECADE_PALETTE(i / max(n_venues - 1, 1))

print('Venue colors:')
for vkey, c in venue_colors.items():
    print(f'  {vkey}: {c}')

for col, ylabel in METRICS:
    if col not in shared_ylims:
        continue

    y_lim = shared_ylims[col]
    xt = shared_xticks.get(col)
    xl = shared_xlim.get(col)

    for vkey in venue_keys_ordered:
        vlabel = multi_data[vkey]['label']
        color = venue_colors[vkey]
        m = multi_data[vkey]['metrics'].get(col)
        if m is None:
            continue

        t_dec = m['t_dec']
        c_dec = m['c_dec']

        # Individual figure
        ax_w, ax_h = 3, 1.2
        lm, bm, rm, tm = 0.85, 0.55, 0.15, 0.30
        fw, fh = lm + ax_w + rm, bm + ax_h + tm
        fig = plt.figure(figsize=(fw, fh), dpi=300)
        ax = fig.add_axes([lm/fw, bm/fh, ax_w/fw, ax_h/fh])

        # Control
        if not c_dec.empty:
            ax.fill_between(c_dec['decade'], c_dec['ci_lo'], c_dec['ci_hi'],
                            color='darkgray', alpha=0.2)
            ax.plot(c_dec['decade'], c_dec['mean'],
                    '-o', color='darkgray', linewidth=1.8, alpha=0.7,
                    markersize=4, label='Control')

        # Treated
        if not t_dec.empty:
            ax.fill_between(t_dec['decade'], t_dec['ci_lo'], t_dec['ci_hi'],
                            color=color, alpha=0.2)
            ax.plot(t_dec['decade'], t_dec['mean'],
                    '-o', color=color, markersize=4, linewidth=1.8, alpha=0.8,
                    label='Treated')

        ax.set_xlabel('Decade', fontsize=8)
        ax.set_ylabel(ylabel, fontsize=8)
        ax.tick_params(labelsize=6)
        # ax.set_title(vlabel, fontsize=9, pad=4)

        # Shared axes across venues for this metric
        ax.set_ylim(y_lim)
        if xt is not None:
            ax.set_xticks(xt)
        if xl is not None:
            ax.set_xlim(xl)

        # Save
        out_path = OUTPUT_DIR / f'multi_{vkey}_{col}_by_decade.png'
        if SAVE_FIGS:
            out_path.parent.mkdir(parents=True, exist_ok=True)
            fig.savefig(out_path, dpi=600, bbox_inches='tight', transparent=True)
            fig.savefig(out_path.with_suffix('.pdf'), bbox_inches='tight', transparent=True)
            print(f'  Saved → {out_path}')
            plt.close(fig)
        else:
            plt.show()
```
